```python
#Calculator Program

def add():
  a=int(input("Enter the first number:"))
  b=int(input("Enter the second number:"))
  sum=a+b
  print("Sum=",sum)
def subtract():
  a=int(input("Enter the first number:"))
  b=int(input("Enter the second number:"))
  diff=a-b
  print("Difference=",diff)
def multiply():
  a=int(input("Enter the first number:"))
  b=int(input("Enter the second number:"))
  pro=a*b
  print("Product=",pro)
def divide():
  a=int(input("Enter the first number:"))
  b=int(input("Enter the second number:"))
  quot=a/b
  print("Quotient=",quot)

print("1.Add")
print("2.Subtract")
print("3.Multiply")
print("4.Divide")
c=int(input("Enter your choice:"))
if c==1:
  add()
elif c==2:
  subtract()
elif c==3:
  multiply()
elif c==4:
  divide()
else:
  print("Enter another choice")
```
